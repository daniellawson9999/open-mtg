```python
# coding: utf-8

# In[1]:

import mcts
from deck import *
from game import *
from player import *
from phases import Phases

player_0_wins = 0
player_1_wins = 0
prev_phase = ""
printed_attackers_blockers = False
attackers = False
        #self.phases = ["Main Phase", "Declare Attackers Step", "Declare Blockers Step", "509.2", "510.1c",
#                       "Combat Damage Step", "Main Phase", "End Step"]
for i in range(1):
    game = Game([Player(get_8ed_core_gold_deck()), Player(get_8ed_core_silver_deck())])
    game.start_game()
    if game.active_player.index == 0:
        print("Gold player's turn")
    else:
        print("Silver player's turn")
    while not game.is_over():
        if game.player_with_priority.index is 1:
            move = game.player_with_priority.determine_move(method="random", game=game)
        else:
            #            move = game.player_with_priority.determine_move(method="random", game=game)
            if len(game.get_moves()) == 1:
                move = game.get_moves()[0]
            else:
                move = mcts.uct(game, itermax=5)
                
        game.make_move(move, True)
        
        #current_phase = game.phases[game.current_phase_index]
        current_phase = game.current_phase_index
        #if current_phase == "510.1c" and not printed_attackers_blockers:
        if current_phase == Phases.COMBAT_DAMAGE_STEP_510_1c and not printed_attackers_blockers:
            printed_attackers_blockers = True
            if len(game.attackers)>0:
                attackers = True
                print("  current phase is: %s" % (current_phase))
                print("    attacking creatures %s: " % (game.attackers))
            if len(game.blockers)>0:
                print("    blocking creatures %s: " % (game.blockers))
         
        #if current_phase == "Combat Damage Step" and attackers:
        if current_phase == Phases.COMBAT_DAMAGE_STEP and attackers:
            print("      Gold player's life: %i, Silver player's life: %i" %(game.players[0].life, game.players[1].life))
        #if current_phase == "End Step":
        if current_phase == Phases.END_STEP:
            attackers = False
            printed_attackers_blockers = False
            gold_permanents = []
            silver_permanents = []
            for permanent in game.battlefield:
                if permanent.owner.index == 0:
                    gold_permanents.append(permanent)
                else:
                    silver_permanents.append(permanent)
            print("  current phase is: %s" % (current_phase))            
            if game.active_player.index == 0:
                print("    gold player's permanents: %s" % (gold_permanents))
                print("\nSilver player's turn ")
            else:
                
                print("    silver player's permanents: %s" % (silver_permanents))
                print("\nGold player's turn " )
        prev_phase = current_phase
    
    
    if game.players[1].has_lost:
        player_0_wins += 1
    elif game.players[0].has_lost:
        player_1_wins += 1
    print("game is over! current standings: %i - %i" % (player_0_wins, player_1_wins))
print("player 0 wins: %i, player 1 wins: %i" % (player_0_wins, player_1_wins))
```
